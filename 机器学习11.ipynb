# AdaBoost&图像识别

Bagging算法的特征
1. 在做重复抽样的时候，每一条数据都是平等的
2. 每一个决策树的投票权都是平等的

后一个模型都是针对前一个模型的错误而做的专项提升。AdaBoost

**增加出错数据的权重，通过使模型更多关注错误数据来改善模型**

**alpha**：可信程度

反复调整、反复训练直到新模型的准确性低于50%，即alpha小于0；或者认为模型数量足够多。
