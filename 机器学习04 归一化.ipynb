 # 归一化
 
 ## 最大最小值归一化
 
 Method = (x-min)/(max-min)
 1. 从preprocessing子模块导入MinMaxScaler类
 2. 创建对象
 3. 调用fit_transform
 4. 将转换后的二维数作为特征集交给机器学习模型进行训练
 5. 将带预测的数据集与训练集整合后归一化
 
还存在标准化，正则化等数据规范化处理

```python
import pandas as pd
df = pd.read_excel('pokemon_knn.xlsx')
df
```

```python
df_unknown = pd.read_excel('pokemon_knn.xlsx',sheet_name=1)
df_unknown['攻击']=150
df_unknown['防御']=50
df_unknown
```

```python
# 交给mms的数据应该是纯数据
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
mms.fit_transform(df.iloc[:,2:21])
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

knn = KNeighborsClassifier()

gscv = GridSearchCV(knn, {'n_neighbors':range(1,10,2)}, cv=5)

df_train, df_test = train_test_split(df)

x = mms.fit_transform(df_train.iloc[:,2:22])
y = df_train['主分类']

gscv.fit(x,y)
pd.DataFrame( gscv.cv_results_ )
```

```python
x = mms.fit_transform(df.iloc[:,2:22])
y = df['主分类']

model = knn.fit(x,y)

model.predict(df_unknown.iloc[:,2:22])
```

```python
x_unknown = mms.fit_transform(df.append(df_unknown).iloc[:,2:22])[-1:,:]
```
