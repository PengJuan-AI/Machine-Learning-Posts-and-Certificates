```python
import pandas as pd
df = pd.read_excel('房源分类-5月.xlsx')
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid',{'font.sans-serif':['simhei','Arial']})
```

```python
sns.stripplot(data=df, x='面积',y='分类')
```

```python
def housetype(s):
    if s<100:
        res = 'Economic'
    elif s<140:
        res = 'Regular'
    elif s<200:
        res = 'Upgraded'
    else: 
        res = 'Luxury'
    return res
            
```

```python
# 使用apply将这个函数运用于dataframe的每一行
df2 = pd.read_excel('待分类房源.xlsx')
df2.head()
```

```python
df2['分类'] = df2.apply( lambda x: housetype(x['面积']), axis=1)
df2
```

逐步将知识抽象化，得到一个普适的判断标准，然后编写可以应用于不同数据集的代码
