机器学习



1. 明确问题：分类/回归/聚类


如果要完成分类和回归问题，事先需要一组已经标记好的训练数据——有监督学习
聚类问题（不知道要分为几种类别）——无监督学习

2. 数据预处理

数据清洗：garbage in, garbage out. 活用pandas进行数据处理

3. 特征工程

* 选择合适的特征用于预测目标。
* 处理特征之间的相关性
* 创建新的特征值
* 对特征值进行编码和统一化（虚拟变量）

数据降维：将原有数据集的特征只降为更少的特征（原数据维度太高过于复杂）

4. 选择算法

5. 模型训练

对整个数据集进行计算，得到一套具体的训练规则。很多时候考验软硬件的能力。

6. 评估优化

用真实数据检验模型。
