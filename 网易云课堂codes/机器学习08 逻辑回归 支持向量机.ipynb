# 逻辑回归

使用linear_model中的LogisticRegression类

常用参数
* penalty：惩罚项，可选值为l1或l2
* C：正则化系数倒数，越小代表越强的正则化，默认为1
* solver： 优化算法选择

```python
import pandas as pd
df = pd.read_csv('心脏疾病数据集.csv')
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

df_train, df_test = train_test_split(df, test_size=0.3, random_state=42)

model = LogisticRegression()

model.fit( df_train.iloc[:,:-1], df_train['诊断结果'])

y_predict = model.predict(df_test.iloc[:,:-1])

#将测试集真实诊断结果与model的预测结果进行比较评分，输出各类分值
print('Accuracy: ', accuracy_score(df_test['诊断结果'], y_predict))

```

# 支持向量机

选择一条分割线，使得两种类别相距对方的距离最远。即，怎样找到间距最大的两条平行边界线。

硬间隔：通道之间没有其他的点

软间隔：通道之间允许存在分类错误的点

因此，算法的目标即，找一个软间隔最大的通道





在支持向量机里，不管使用多少指标，多少坐标维度，我们都把划分类编的几何形状称为**超平面**。

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

df_train, df_test = train_test_split(df, test_size=0.3, random_state=42)

model = SVC()

model.fit( df_train.iloc[:,:-1], df_train['诊断结果'])

y_predict = model.predict(df_test.iloc[:,:-1])

#将测试集真实诊断结果与model的预测结果进行比较评分，输出各类分值
print('Accuracy: ', accuracy_score(df_test['诊断结果'], y_predict))

```
